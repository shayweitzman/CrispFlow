exploratory data analysis is an approach of analyzing data sets to summarize their main characteristics. <br>
A statistical model can be used or not, but primarily EDA is for seeing what the data can tell us beyond the formal modeling or hypothesis testing task. <br>
we going to use matplotlib and seaborn to present the data and we will try to understand extra information about our data.

```python
import numpy as np -
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import seaborn as sns

warnings.filterwarnings('ignore')
```

## First view on categorial columns

```python
data.describe()
```

```python
cat_col = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']

sns.set_theme()
fig, ax = plt.subplots(nrows=9, ncols=1, figsize=(15, 50))
fig.subplots_adjust(hspace=.5)
plt.title('Countplot for categorical variables')

for i, col in enumerate(cat_col):
    sns.countplot(y=col, data=data, ax = ax[i])
    ax[i].set_title('count of {}'.format(col))
```

#### Takeaway:

- job column: job column contains not only the category but also the status of employement. It maybe worthy to separate them.<br>
- education column : same as job column , we can separate the examples that have any education and not.
- month column: december and may had especially low and high counts, respectively.<br>
- default, 'housing' and 'loan' columns are useful columns to be converted to a binary variables that only contains 0 or 1 as possible values. <br>

## First view on numeric columns

```python
num_col = ['age', 'balance', 'day', 'campaign', 'previous'] # numeric columns

fig, ax = plt.subplots(nrows=5, ncols=1, figsize=(15, 45)) # create subplots
fig.subplots_adjust(hspace=.5) # adjust space between plots

for i, col in enumerate(num_col):
    sns.boxplot(x=data[col], ax=ax[i])
    ax[i].set_title('Distribution of {}'.format(col))
```

#### Takeaway:

- age column: most of age values is distributed between 18 to 70, with median value at 39.
- balance column: balance varied widely, but most of the value was concentrated near 0.
- campaign column : most of the values arround 1-3.
- previous column: mostly zero examples were made previously.

## Correlation Matrix
Correlation can be done only on numeric values, so we transform the categorails features to numeric.

```python
from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
df = data.copy()
for i in df.columns:
    if i not in num_col:
        df[i] = number.fit_transform(df[i])
df.head()
```

```python
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn',linewidths=0.3) #data.corr()-->correlation matrix
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.show()
```

### Present coloums with >0.1 correlation

```python
df1 = df[['age','marital','housing','month','contact','duration','class', 'previous', 'poutcome']]
sns.heatmap(df1.corr(),annot=True,cmap='RdYlGn',linewidths=0.2,annot_kws={'size':20})
fig=plt.gcf()
fig.set_size_inches(13,11)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.show()
```

#### We can see that there is positive association between:
housing -> month,housing -> contact, month->contact,contact -> poutcome, duration->class.<br>
positive association (X->Y) means : when X values increase Y value increase.<br>

#### negative association between:<br>
age->marital, age->housing,housing->poutcome,contact->class,contact->previous, previous->poutcome.<br>
negative association (X->Y) means : when X values increase Y value decrease.<br>

## Class & Age

## Column that effects balance

```python
df = data.copy()
means = []
values = ['no','yes']
means.append(df.query("`default` == 'no'")['balance'].mean())
means.append(df.query("`default` == 'yes'")['balance'].mean())



plt.bar(values, means)
plt.title('Default X balance\n')
plt.xlabel('default')
plt.ylabel('balance')
```

```python
df = data.copy()
means = []
values = ['no','yes']
means.append(df.query("`loan` == 'no'")['balance'].mean())
means.append(df.query("`loan` == 'yes'")['balance'].mean())



plt.bar(values, means)
plt.title('balance X loan\n')
plt.xlabel('loan')
plt.ylabel('balance')
```

```python
df = data.copy()
means = []
values = ['married','singleyes', 'divorced']
means.append(df.query("`marital` == 'married'")['balance'].mean())
means.append(df.query("`marital` == 'single'")['balance'].mean())
means.append(df.query("`marital` == 'divorced'")['balance'].mean())



plt.bar(values, means)
plt.title('Marital X balance\n')
plt.xlabel('marital')
plt.ylabel('balance')
```

#### Takeaways:
- The value of 'default' col has major effect on balance value.
- Balance affected also by loan value.
- Divorced have the lowest balance mean.
